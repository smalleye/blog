```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the square wave
T = 2 * np.pi  # Period of the square wave
x = np.linspace(0, T, 1000)  # x-axis values
num_terms = 3  # Number of Fourier terms

# Function to compute the Fourier series approximation
def square_wave_fourier(x, terms):
    approximation = np.zeros_like(x)
    for n in range(1, 2 * terms, 2):  # Only odd terms (1, 3, 5, ...)
        approximation += (4 / (n * np.pi)) * np.sin(n * x)
    return approximation

# Generate the Fourier series approximation
approximation = square_wave_fourier(x, num_terms)

# Plot the square wave and its Fourier approximation
plt.figure(figsize=(10, 6))
plt.plot(x, approximation, label=f'Fourier Approximation ({num_terms} terms)', color='blue')
plt.axhline(1, color='red', linestyle='--', label='Ideal Square Wave')
plt.axhline(-1, color='red', linestyle='--')
plt.title('Square Wave Approximation using Fourier Series')
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()
plt.show()
```
